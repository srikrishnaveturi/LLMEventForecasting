```python
import pandas as pd
from tqdm import tqdm
import requests
```

```python
finalData = pd.DataFrame({
    'question': [], 
    'qtype': [],
    'answer': [],
    'resolve_time': [],
    'active_state': []
})
nxt = 'https://www.metaculus.com/api2/questions/?limit=100'
for counter in tqdm(range(100)):
    response = requests.get(nxt)
    nxt = response.json()['next']
    for res1 in response.json()['results']:
        if res1['active_state'] == 'RESOLVED':
            finalData.loc[len(finalData.index)] = [res1['title'], res1['possibilities'], res1['resolution'], res1['resolve_time'], res1['active_state']]
    if nxt is None:
        break
    
```
