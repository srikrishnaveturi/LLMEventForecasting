```python
allData = pd.read_csv('data/CombinedData.csv')
allData
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(list(allData['Year']))
plt.show()
```

```python
labels = {
    'Yes': 0,
     'No': 0,
     'MCQ': 0,
     'Num': 0
}
for i in range(len(allData['Answer'])):
    ans = allData['Answer'][i]
    if ans == 'yes':
        labels['Yes'] += 1
    elif ans == 'no':
        labels['No'] += 1
    elif allData['QType'][i] == 'mc':
        labels['MCQ'] += 1
    else:
        labels['Num'] += 1
```

```python
keys = list(labels.keys())
vals = [labels[key] for key in keys]

plt.pie(vals, labels=keys, autopct='%1.1f%%')
plt.show()
```
